# Example of testing a number is prime or not

```python
def is_prime(num):
    '''
    Method to check the number is prime or not
    '''
    
    for n in range(2,num):
        if num % n == 0:
            print('This number is not Prime ',num)
            break;
        else:
            print(num, 'is a prime number !')
```
